# Splitting data into features and target variables

```python
x = train_df.iloc[:, :train_df.columns.get_loc("starch_water")+1]
y_dissolution = train_df.iloc[:, train_df.columns.get_loc("dissolution_av"):train_df.columns.get_loc("dissolution_av")+1]
y_residual = train_df.iloc[:, train_df.columns.get_loc("resodual_solvent"):train_df.columns.get_loc("resodual_solvent")+1]
y_impurity = train_df.iloc[:, train_df.columns.get_loc("impurities_total"):train_df.columns.get_loc("impurities_total")+1]
```

# Creating machine learning models


```python
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import GradientBoostingRegressor
gbr_dissolution = GradientBoostingRegressor()
gbr_dissolution.fit(x, y_dissolution.values.ravel())
gbr_residual = GradientBoostingRegressor()
gbr_residual.fit(x, y_residual.values.ravel())
br_impurity = BaggingRegressor()
br_impurity.fit(x, y_impurity.values.ravel())
```

# Defining callable functions for the Anvil server

```python
updates = []
```

```python
@anvil.server.callable
def predict_impurity(params):
  prediction1 = gbr_dissolution.predict([params[1:]])
  prediction2 = gbr_residual.predict([params[1:]])
  prediction3 = br_impurity.predict([params[1:]])
  if params not in updates:
    updates.append(params)
  return [prediction1[0], prediction2[0], prediction3[0]]
```

```python
@anvil.server.callable
def upload_data(params):
  global updates
  global train_df
  global gbr_dissolution
  global gbr_residual
  global br_impurity
  for data in updates:
    if data[0] == params[0]:
      data_copy = data.copy()
      new_row = data_copy[1:] + params[1:]
      row1 = pd.Series(new_row, index=train_df.columns)
      train_df = train_df.append(row1,ignore_index=True)
      updates.remove(data)

      x = train_df.iloc[:, :train_df.columns.get_loc("starch_water")+1]
      y_dissolution = train_df.iloc[:, train_df.columns.get_loc("dissolution_av"):train_df.columns.get_loc("dissolution_av")+1]
      y_residual = train_df.iloc[:, train_df.columns.get_loc("resodual_solvent"):train_df.columns.get_loc("resodual_solvent")+1]
      y_impurity = train_df.iloc[:, train_df.columns.get_loc("impurities_total"):train_df.columns.get_loc("impurities_total")+1]

      gbr_dissolution.fit(x, y_dissolution.values.ravel())
      gbr_residual.fit(x, y_residual.values.ravel())
      br_impurity.fit(x, y_impurity.values.ravel())

      return "Data recorded and model trained successfully"
  return "Error"
```
